```python
import matplotlib

import matplotlib.pyplot as plt
import numpy as np
from os.path import isfile, isdir
import scipy.io as sio

# Enable interactive figures for latest versions of Jupyter Lab, with 'jupyter-matplotlib' 'widgetsnbextension' installed and activated.
#%matplotlib widget
# If no figure is showing, replace with
%matplotlib inline

from mpl_toolkits.mplot3d import Axes3D
```

```python
_data_mat = sio.loadmat('data/KS.mat')
Data = _data_mat['Data_KS']
Data_test = _data_mat['Data_KS_test']
```

```python
U, S, Vh = np.linalg.svd(Data, full_matrices=False)
V = Vh.T
S_norm = S / sum(S)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
ax1.grid()
ax1.set_title('Normalized singular values')
ax1.set_xlabel('Singular Value #')
ax1.scatter(range(1, len(S_norm)+1), S_norm)
ax1.set_xlim((0, 80))

ax2.grid()
ax2.set_title('Cumulative energy')
ax2.set_xlabel('Singular Value #')
ax2.scatter(range(1, len(S_norm)+1), np.cumsum(S_norm))
ax2.set_xlim((0, 80))
plt.savefig('figs/KS_singular_values.png')
```

```python
#%% Reduce dimensionality
# Truncation at r = 50
r = 50
Ur = U[:, 0:r]
Sigmar = np.diag(S[0:r])
Vr = V[:, 0:r]
Data_r = Ur @ Sigmar @ Vr.T.conj()

xx, yy = np.meshgrid(np.linspace(0, 2 * np.pi *  99 / Data_test.shape[0], 100), 
                     np.linspace(0, 99 * 0.01, 100))
fig = plt.figure()
ax3 = plt.axes(projection='3d')
ax3.plot_surface(xx, yy, Data[:100, :100].T)
ax3.set_title('Real data')
ax3.set_xlabel('Position')
ax3.set_ylabel('Time [s]')
ax3.view_init(30, -10)
plt.savefig('figs/KS_real_data.png')

plt.figure()
ax3 = plt.axes(projection='3d')
ax3.plot_surface(xx, yy, Data_r[:100, :100].T)
ax3.view_init(30, -10)
ax3.set_title('Reconstructed data')
ax3.set_xlabel('Position')
ax3.set_ylabel('Time [s]')
plt.savefig('figs/KS_truncation_reconstruction.png')
```

```python
# Project the data into the low-rank feature space
Data_lowdim = Ur.T.conj() @ Data
# 87.5% of the data is used for training, the remainder is used for validation
train_part = 0.875
train_samples = round(train_part * Data_lowdim.shape[1])
Data_train = Data_lowdim[:, 0:train_samples]
Data_val = Data_lowdim[:, train_samples:]

# Normalize the data
norm_mean = np.mean(Data_train, axis=1).reshape((-1, 1))
norm_std = np.std(Data_train, axis=1).reshape((-1, 1))
Data_train_norm = (Data_train - norm_mean) / norm_std
Data_val_norm = (Data_val - norm_mean) / norm_std
```

```python
#%% Construct the training and validation sets
tau = 0  # Embedded time-delays

# REMARK: we embedd the time-delays in the low-rank subspace, and then build the training and validation data matrices
X_train = np.empty(((tau + 1) * Data_train_norm.shape[0], 0))
Y_train = np.empty((Data_train_norm.shape[0], 0))
X_val = np.empty(((tau + 1) * Data_train_norm.shape[0], 0))
Y_val = np.empty((Data_train_norm.shape[0], 0))

for i in range(tau + 1, Data_train_norm.shape[1] - 1):
    X_train = np.append(X_train, Data_train_norm[:, i - tau:i + 1].T.reshape(-1, 1), axis=1)
    Y_train = np.append(Y_train, Data_train_norm[:, i + 1].T.reshape(-1, 1), axis=1)

for i in range(tau, Data_val_norm.shape[1] - 1):
    X_val = np.append(X_val, Data_val_norm[:, i - tau:i + 1].T.reshape(-1, 1), axis=1)
    Y_val = np.append(Y_val, Data_val_norm[:, i + 1].T.reshape(-1, 1), axis=1)
```

```python
#%% Neural network training
import tensorflow as tf
from tensorflow.keras import Model
from tensorflow.keras.layers import Dense, Input, LeakyReLU, Add
from tensorflow.keras.callbacks import EarlyStopping
import os
from pathlib import Path

_nn_path = str(Path('saved/trained_network_part2a'))  # This ensures robustness wrt the OS

if not tf.__version__.startswith('2'):
    tf.enable_eager_execution()

if isfile(_nn_path) or isdir(_nn_path):
    nn = tf.keras.models.load_model(_nn_path)
    print('Pre-loaded NN model imported')
else:
    # To manually stop the training, put a file named STOP.txt in the main directory
    class CatchUserInterruptCallback(tf.keras.callbacks.Callback):
        def on_epoch_end(self, epoch, logs=None):
            if os.isfile('STOP.txt'):
                print('Stopping on user request...')
                self.stopped_epoch = epoch
                self.model.stop_training = True
                os.remove('STOP.txt')

    # Autoencoder structure
    x0 = Input(shape=(X_train.shape[0],))
    x1 = Dense(units=64, activation='tanh')(x0)
    x2 = LeakyReLU(alpha=0.1)(Dense(units=32)(x1))
    x3 = Dense(units=24, activation='tanh')(x2)
    x4 = Dense(units=12, activation='tanh')(x3)
    x6 = Dense(units=12, activation='tanh')(x4)
    x7 = Dense(units=24, activation='tanh')(x6)
    x8 = LeakyReLU(alpha=0.1)(Dense(units=32)(x7))
    x9 = Dense(units=64, activation='tanh')(x8)
    x10 = Dense(units=Y_train.shape[0], activation='tanh')(x9)
    x11 = Add()([x0, x10])

    nn = Model(x0, x11)
    nn.compile(tf.keras.optimizers.Adam(1e-3), loss='mse')

    # Early stopping rule based on validation loss
    esr = EarlyStopping(monitor='val_loss', verbose=1, restore_best_weights=True, patience=1000)
    nn.summary()
    nn.fit(X_train.T, Y_train.T, validation_data=(X_val.T, Y_val.T),
           epochs=5000, batch_size=128, shuffle=True, callbacks=[CatchUserInterruptCallback(), esr])

    nn.save(_nn_path)
    print("Neural network trained and dumped to file")
```

```python
# Project the test data-set into the low-dimensional features space
Data_test_lowdim = Ur.T.conj() @ Data_test

# Standardization of the initial conditions
Data_nn = (Data_test_lowdim[:, 0:tau + 1] - norm_mean) / norm_std
x0 = Data_nn.T.reshape(-1, 1)

# Open-loop simulation of the NN with the given initial conditions
for t in range(tau, Data_test.shape[1] - 1):
    _y_nn = nn.predict(x0.T)
    Data_nn = np.append(Data_nn, _y_nn.reshape(-1, 1), axis=1)
    x0 = Data_nn[:, -tau - 1:].T.reshape(-1, 1)

# De-standardization and projection into the "full" space
Data_nn_denorm = Data_nn * norm_std + norm_mean
Data_nn_full = Ur @ Data_nn_denorm
```

```python
xx, yy = np.meshgrid(np.linspace(0, 2 * np.pi, Data_test.shape[0]),
                     np.linspace(0, (Data_test.shape[1]-1) * 0.01, Data_test.shape[1]))
fig = plt.figure()
ax3 = plt.axes(projection='3d')
ax3.plot_surface(xx, yy, Data_test.T)
ax3.plot_surface(xx, yy, Data_nn_full.T)
ax3.set_title('Test')
ax3.set_ylabel('Time')
ax3.set_xlabel('Position')
plt.savefig('figs/KS_NN_performance_3d.png')
```

```python
fig = plt.figure()
ax3 = plt.axes(projection='3d')
ax3.plot_surface(xx, yy, Data_test.T - Data_nn_full.T)
ax3.set_title('Error')
ax3.set_ylabel('Time')
ax3.set_xlabel('Position')
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(13, 5))
map1 = ax1.pcolor(xx, yy, Data_test.T, vmin=-75, vmax=75)
ax1.set_title('Test data')
ax1.set_ylabel('Time [s]')
ax1.set_xlabel('Position')
map2 = ax2.pcolor(xx, yy, Data_nn_full.T, vmin=-75, vmax=75)
ax2.set_title('NN prediction')
ax2.set_ylabel('Time [s]')
ax2.set_xlabel('Position')

fig.subplots_adjust(right=0.85)
cbar_ax = fig.add_axes([0.875, 0.125, 0.03, 0.75])
fig.colorbar(map2, cax=cbar_ax)
plt.savefig('figs/KS_NN_performance.png')
```
