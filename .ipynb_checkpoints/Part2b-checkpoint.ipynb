```python
import time
import matplotlib
from matplotlib import animation

import matplotlib.pyplot as plt
import numpy as np

dT = 0.05

Data_flat = np.genfromtxt('reaction_diffusion.csv', delimiter=',')
Data_tensor = Data_flat.reshape((512, 512, -1), order='F')      # Fortran order as MATLAB
X = Data_flat[:, 0:-1]
Xprime = Data_flat[:, 1:]
```

```python
U, S, Vh = np.linalg.svd(X, full_matrices=False)
V = Vh.T
S_norm = S**2 / sum(S**2)

plt.figure(1, figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.grid()
plt.title('Normalized singular values')
plt.xlabel('Singular Value #')
plt.scatter(range(0, len(S_norm)), S_norm)

plt.subplot(1, 2, 2)
plt.grid()
plt.title('Cumulative energy')
plt.xlabel('Singular Value #')
plt.scatter(range(0, len(S_norm)), np.cumsum(S_norm))
plt.show()
```

```python
r = 10
Ur = U[:, 0:r]
Sigmar = np.diag(S[0:r])
Vr = V[:, 0:r]
# Reconstruct the data via truncated SVD to assess the goodness
X_r = Ur @ Sigmar @ Vr.T.conj()
Data_r_tensor = X_r.reshape((512, 512, -1), order='F')

vmin = Data_flat.flatten().min()
vmax = Data_flat.flatten().max()
```

```python
xx, yy = np.meshgrid(range(0, 512), range(0, 512))
fig, (ax1, ax2) = plt.subplots(1, 2)
fig.suptitle('Reaction-diffusion system')
map1 = ax1.imshow(Data_tensor[:, :, 0], vmin=vmin, vmax=vmax, origin='lower')
ax1.set_title('Real data')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
map2 = ax2.imshow(Data_r_tensor[:, :, 0], vmin=vmin, vmax=vmax, origin='lower')
ax2.set_title('Reconstructed data')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
# fig.colorbar(map2)

# function to update figure
def updatefig(j):
    fig.get_axes()[0].get_images()[0].set_data(Data_tensor[:, :, j])
    fig.get_axes()[1].get_images()[0].set_data(Data_r_tensor[:, :, j])
    return fig

ani = animation.FuncAnimation(fig, updatefig, frames=range(Data_tensor.shape[2]-1), interval=0.25, blit=False, repeat=False)
```

```python
A_tilde = Ur.conj().T @ Xprime @ Vr @ np.linalg.inv(Sigmar)

Lambda, W = np.linalg.eig(A_tilde)
Phi = Xprime @ Vr @ np.linalg.inv(Sigmar) @ W
omega = np.log(Lambda) / dT
B = np.linalg.pinv(Phi) @ X[:, 0]
```

```python
from sklearn.metrics import mean_squared_error

plt.figure(4, figsize=(5, 5))
plt.scatter(np.real(omega), np.imag(omega))
plt.title('DMD Eigenvalues')
plt.grid()
plt.show()

t = np.linspace(0, dT * Data_tensor.shape[2], Data_tensor.shape[2])

Data_pred_flat = np.real([sum([Phi[:, k] * np.exp(omega[k] * tau) * B[k] for k in range(0, B.shape[0])]) for tau in t]).T

MSE_flat = np.square(Data_flat - Data_pred_flat).mean(axis=0)

plt.figure(4, figsize=(5, 5))
plt.plot(t, MSE_flat)
plt.xlabel('t [s]')
plt.ylabel('MSE')
plt.title('DMD prediction MSE')
plt.grid()
plt.show()
```
